### 문항.2

```python
df2 = pd.read_csv('../04.지도시각화/data/경기도_주민등록인구집계현황.csv',encoding='euc-kr')
df2.head(3)
```

```python
df2 = df2[['행정구역구분명','행정구역명','총 인구수']]
df2.head(3)
```

```python
gu_city = ['수원시', '고양시', '용인시', '성남시', '안산시', '안양시']
df2.행정구역명 = df2.행정구역명.str.strip()
drop_index = []
for index in df2[df2.행정구역구분명 == '시군'].index:
    if df2.행정구역명[index].split()[-1] in gu_city:
        drop_index.append(index)
df2.drop(drop_index, inplace=True)
df2.head()
```

```python
df2.drop(0)
```

```python
df2 = df2[(df2.행정구역구분명 == '시군') | (df2.행정구역구분명 == '구')]
df2.head()
```

```python
df2['시군구'] = df2.행정구역명.apply(lambda x: ''.join(x.split()[1]))
df2.head()
```

```python
df2 = df2.set_index('시군구')
df2.head()
```

```python
sort_df = df[['공원면적','시군구']].groupby('시군구').sum()
sort_df.head()
```

```python
sort_df2=df2.sort_index()
sort_df2.head()
```

```python
df3 = pd.merge(left = sort_df2 , right = sort_df, how = "right", on = "시군구")
df3.head()
```

```python
df3['인당 공원면적'] = df3['공원면적']/df3['총 인구수']
df3.head()
```
