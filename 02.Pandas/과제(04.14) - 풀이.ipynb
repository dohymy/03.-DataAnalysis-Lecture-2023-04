```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1) 데이터 파일 읽기
filename = 'data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv'
columns = '구분,총계,2013년 이전 설치된 CCTV,2013년,2014년,2015년,2016년,2017년,2018년,2019년,2020년,2021년,2022년'.split(',')
cctv = pd.read_csv(filename, encoding='euc-kr', thousands=',')
cctv = cctv[columns]
cctv.head()
```

```python
# 2) null 데이터를 먼저 잡는다
cctv.isna().sum().sum()
```

```python
# 3) 최근 증가율
cctv.loc[:,'2020년':'2022년'].sum(axis=1).head()
```

```python
cctv['최근 증가율'] = cctv.iloc[:,10:].sum(axis=1) / cctv.iloc[:,2:10].sum(axis=1)
cctv['최근 증가율'] = (cctv['최근 증가율'] * 100).round(2)
cctv.head()
```

```python
# 4) Selection and Rename
cctv = cctv[['구분', '총계', '최근 증가율']]
cctv.head()
```

```python
cctv.rename(columns={'구분':'구별', '총계':'CCTV 댓수'}, inplace=True)
cctv.head()
```

```python
# 5) Filtering
cctv.drop([0], inplace=True)
cctv.head()
```

```python
# 6) 구 이름에서 공백 지우기
cctv.구별 = cctv.구별.str.replace(' ','')
cctv.head(3)
```

```python
# 7) csv파일에 저장
cctv.to_csv('data/cctv2.csv', index=False)
df = pd.read_csv('data/cctv2.csv')
df.head()
```
