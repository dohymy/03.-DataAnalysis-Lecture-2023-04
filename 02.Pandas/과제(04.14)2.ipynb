```python
import numpy as np
import pandas as pd
```

```python
filename = 'data/서울시인구.txt'
pop = pd.read_csv(filename, sep='\t', thousands=',')
pop.head()
```

```python
del pop['기간'], pop['세대'], pop['인구.1'], pop['인구.2'], pop['인구.4'], pop['인구.5'], pop['인구.7'], pop['인구.8'], pop['세대당인구']
pop.head()
```

```python
pop.drop([0, 1, 2],inplace=True)
pop.head()
```

```python
columns = '자치구,인구,인구.3,인구.6,65세이상고령자'.split(',')
for col in columns[1:]:
    pop[col] = pop[col].str.replace(',','').astype(int)
pop.head(3)
```

```python
pop['외국인비율'] = (pop['인구.6']/pop['인구']*100).round(2)
pop.head()
```

```python
pop['고령자비율'] = (pop['65세이상고령자']/pop['인구']*100).round(2)
pop.head(3)
```

```python
pop.columns='구별', '인구수', '내국인', '외국인', '고령자', '외국인비율', '고령자비율'
pop.head(3)
```

```python
pop.set_index('구별', inplace=True)
pop.head()
```

```python
pop.reset_index(inplace=True)
pop.head(3)
```

```python
pop.to_csv('data/pop.csv')
```
