### 인구소멸 위기 지역

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/서울시인구 copy.txt')
df.head()
```

```python
df = pd.read_csv('data/주민등록인구집계현황.csv', encoding='euc-kr')
df.head()
```

```python
df.isna().sum().sum()
```

```python
df['가임기여성인구수'] = df['20~29세 (여)'] + df['30~39세 (여)']
df.head()
```

```python
df['60세이상노령인구'] = df['60~69세'] + df['70~79세'] + df['80~89세'] + df['90~99세'] + df['100세 이상']
df.head()
```

```python
df['인구소멸위기'] = df['가임기여성인구수'] / df['60세이상노령인구']
df.head()
```

```python
df = df[(df.행정구역구분명 == '시군') | (df.행정구역구분명 == '구')]
df.head()
```

```python
df = df[['행정구역구분명', '행정구역명', '가임기여성인구수', '60세이상노령인구', '인구소멸위기']]
df.head()
```

```python
df.sort_values(by='인구소멸위기', ascending=True, inplace=True)
df.head(10)
```
