### 문제.1

```python
import numpy as np
import pandas as pd
```

```python
filename = 'data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv'
cctv = pd.read_csv(filename, encoding='euc-kr')
cctv.head()
```

```python
columns = '구분,총계,2013년 이전 설치된 CCTV,2013년,2014년,2015년,2016년,2017년,2018년,2019년,2020년,2021년,2022년'.split(',')
cctv = cctv[columns]
cctv.head()
```

```python
cctv.drop(0, inplace=True)
cctv.head(3)
```

```python
cctv.rename(columns={"구분": "구별", "총계": "CCTV 댓수"}, inplace=True)
cctv.head(3)
```

```python
for col in columns[2:]:
    cctv[col] = cctv[col].str.replace(',','').astype(int)
cctv.head()
```

```python
cctv['최근 증가율'] = ((cctv['2013년 이전 설치된 CCTV'] + cctv['2013년'] + cctv['2014년'] + cctv['2015년'] + cctv['2016년'] + cctv['2017년'] + cctv['2018년'] + cctv['2019년']) / (cctv['2020년'] + cctv['2021년']+ cctv['2022년']) * 100).round(2)
cctv.head(3)
```

```python
del cctv['2013년 이전 설치된 CCTV'], cctv['2013년'], cctv['2014년'], cctv['2015년'], cctv['2016년'], cctv['2017년'], cctv['2018년'], cctv['2019년'], cctv['2020년'], cctv['2021년'], cctv['2022년']
cctv.head(3)
```

```python
cctv.to_csv('data/cctv.csv')
```
