### 5. Pandas 시각화

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

- 라인 그래프

```python
np.random.seed(2023)
df = pd.DataFrame(np.random.randn(100,3),columns=list('ABC')
                  , index=pd.date_range('2023-04-01', periods=100)).cumsum()
df.head()
```

```python
# 디폴트가 라인그래프
df.plot()
plt.title('Pandas의 plot 메소드')
plt.xlabel('일자'), plt.ylabel('데이터')
plt.show()
```

- 막대 차트

```python
# kind을 쓰는 방식
iris = sns.load_dataset('iris')
# rot : x축 데이터의 기울기
iris.sepal_length[:30].plot(kind='bar', rot=0, figsize=(6,3))
plt.title('Iris Setosa 품종의 sepal length');
```

```python
iris.sepal_length[:30].plot(kind='barh')
plt.title('Iris Setosa 품종의 sepal length');
```

```python
df2 = iris.groupby('species').mean()
df2
```

```python
# bar 메소드를 쓰는 방식
df2.plot.bar(rot=0, figsize=(5,3))
plt.title('Iris 품종별 Feature의 평균');
```

```python
# x/y lim : x/y의 범례(limit) 
df2.columns.name = 'Feature'
df2.plot.bar(rot=0, figsize=(5,3))
plt.ylim(0,10)
plt.title('Iris 품종별 Feature의 평균');
```

- 파이 차트

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%.1f%%')
plt.title('선실별 승객수');
```

- 히스토그램, kernel density

```python
iris.plot.hist()
plt.show()
```

```python
iris.plot.hist()
plt.title('Iris 피쳐 히스토그램')
iris.plot.kde()
plt.title('Iris kde')
plt.show()
```

- 박스 플롯

```python
iris[iris.species == 'setosa'].plot.box()
plt.title('Iris Setosa 품종의 각 피쳐별 박스플롯');
```

```python
iris.boxplot(by='species')
plt.tight_layout()
plt.title('Iris 각 피쳐의 품종별 박스 플롯');
```
