### 1. Matplotlib 소개

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

##### 1. 라인 플롯

```python
plt.plot([0,1,4,9])     # X가 생략되면 [0,1,2,...]
plt.title('라인 플롯')
plt.grid()
```

##### 스타일 지정

```python
# rs-- 스타일
# color: Red, marker: Square, linestyle: --(점선)
plt.plot([0,1,4,9], 'rs--')
plt.title("'rs--' 스타일의 플롯")
plt.grid()
```

```python
plt.plot([0,1,4,9], 'mo-.')
plt.plot([9,4,1,0], 'g^:')
plt.title("'mo-.', 'g^:' 스타일의 플롯")
plt.grid()
```

```python

plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'bo--',
         lw=5,  ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.show()
```

##### 그림 범위

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.grid()
plt.show()
```

- 곡선 그림

```python
X = np.linspace(0, 2*np.pi, 500)
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Sine, Cosine 그래프'), plt.grid()
plt.show()
```

##### Tick 설정

```python
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Tick 설정'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
plt.yticks([-1,0,1])
plt.show()
```

```python
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('LaTex, 문자열로 tick label 정의'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1])
plt.show()
```

- 범례(legend)

```python
plt.plot(X, Ysin, label = 'Sine')
plt.plot(X, Ycos, label = 'Cosine')
plt.title('범례 표시')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1]), plt.grid()
plt.legend(loc='upper center')           # upper center position
plt.show()
```

- X축, Y축 Label

```python
plt.plot(X, Ysin, label = 'Sine')
plt.plot(X, Ycos, label = 'Cosine')
plt.title('X축 Y축 라벨 표시')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1]), plt.grid()
plt.legend(loc=0)
plt.xlabel('시간'), plt.ylabel('진폭')
plt.show()
```

- 그림의 크기 지정

```python
plt.figure(figsize=(10,6))
plt.plot(X, Ysin, label = 'Sine')
plt.plot(X, Ycos, label = 'Cosine')
plt.title('그림 크기 지정')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1]), plt.grid()
plt.legend(loc=0)
plt.xlabel('시간'), plt.ylabel('진폭')
plt.show()
```

- 그림 저장

```python
plt.figure(figsize=(10,6))
plt.plot(X, Ysin, label = 'Sine')
plt.plot(X, Ycos, label = 'Cosine')
plt.title('그림 저장', fontsize=15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1]), plt.grid()
plt.legend(loc=0)
plt.xlabel('시간'), plt.ylabel('진폭')
plt.savefig('data/삼각함수.png')
```
