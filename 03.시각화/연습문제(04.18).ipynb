##### 1. Tips 데이터

- 1. 요일별 팁의 비율
- 2. 인원수별 팁의 비율

```python
print(tips.shape)
tips
```

```python
tips['tip_pct'] = (tips.tip / tips.total_bill * 100).round(2)
tips.tail()
```

```python
tip = tips['tip']
day = tips['day']
plt.figure(figsize=(4,2))
plt.xlabel('요일'), plt.ylabel('팁')
plt.bar(day.values, tip.values)
plt.grid()
plt.show()
```

```python
tip = tips['tip']
size = tips['size']
plt.figure(figsize=(4,2))
plt.xlabel('인원수'), plt.ylabel('팁')
plt.bar(size.values, tip.values)
plt.title('인원수별 팁의 비율')
plt.grid()
plt.show()
```

- 서울시 구별 CCTV댓수
    - index를 sort
    - value를 sort

```python
filename = '../02.Pandas/data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv'
columns = '구분,총계'.split(',')
cctv = pd.read_csv(filename, encoding='euc=kr', thousands=',')
cctv = cctv[columns]
cctv.head()
```

```python
cctv.drop([0], inplace=True)
cctv.set_index('구분', inplace=True)
cctv.head(3)
```

```python
# 1. Unsorted
plt.barh(cctv.index, cctv.총계)
plt.title('자치구별 CCTV 설치대수')
plt.show()
```

```python
# 2. index로 정렬
si = cctv.총계.sort_index(ascending=False)
plt.barh(si.index, si.values)
plt.title('자치구별 CCTV 설치대수(자치구명으로 정렬)')
plt.show()
```

```python
# 3. 설치대수로 정렬
sv = cctv.총계.sort_values()
plt.barh(sv.index, sv.values)
plt.title('자치구별 CCTV 설치대수(설치대수로 정렬)')
plt.show()
```
