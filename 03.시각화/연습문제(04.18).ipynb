```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

##### 1. Tips 데이터

- 1. 요일별 팁의 비율
- 2. 인원수별 팁의 비율

```python
print(tips.shape)
tips
```

```python
tips['tip_pct'] = (tips.tip / tips.total_bill * 100).round(2)
tips.tail()
```

```python
tip = tips['tip']
day = tips['day']
plt.figure(figsize=(4,2))
plt.xlabel('요일'), plt.ylabel('팁')
plt.bar(day.values, tip.values)
plt.grid()
plt.show()
```

```python
tip = tips['tip']
size = tips['size']
plt.figure(figsize=(4,2))
plt.xlabel('인원수'), plt.ylabel('팁')
plt.bar(size.values, tip.values)
plt.title('인원수별 팁의 비율')
plt.grid()
plt.show()
```

- 서울시 구별 CCTV댓수
    - index를 sort
    - value를 sort

```python
filename = 'data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv'
cctv = pd.read_csv(filename, encoding='euc-kr')
cctv.head()
```

```python
cctv.sort_values(by='총계', ascending=False, inplace=True)
cctv.head()
```

```python
s = cctv['총계']
plt.barh(s.index, s.values)
plt.title('서울시 구별 CCTV댓수')
plt.show()
```
