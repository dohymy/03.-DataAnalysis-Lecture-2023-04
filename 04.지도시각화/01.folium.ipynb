### Folium 지도 시각화
- 설치 : pip install folium

```python
import folium
folium.__version__
```

```python
map = folium.Map(location=[37.295,127.045], zoom_start=12)  # open street map
folium.Marker(
    location=[37.2949837, 127.0453902], # wg84계 좌표(위도(latitude),경도(longitude))
    popup=folium.Popup('경기고 수원시 영통구 도청로 30', max_width=200), # 클릭 시 뜨는 팝업
    tooltip='경기도청'  # 마우스 올렸을 때 뜨는 문구
).add_to(map)
map
```

```python
# Stamen Toner 지도
map = folium.Map(location=[37.295,127.045], zoom_start=12, tiles='Stamen Toner')
folium.Marker(
    location=[37.2949837, 127.0453902],
    popup=folium.Popup('경기고 수원시 영통구 도청로 30', max_width=200), # 클릭 시 뜨는 팝업
    tooltip='경기도청'  # 마우스 올렸을 때 뜨는 문구
).add_to(map)
map
```

```python
# 여러 형태의 아이콘
map = folium.Map(location=[37.295,127.045], zoom_start=12)
folium.Marker(
    location=[37.2949837, 127.0453902],
    popup=folium.Popup('경기고 수원시 영통구 도청로 30', max_width=200),
    tooltip='경기도청',
    icon=folium.Icon(
        color='purple',
        icon_color='white',
        icon='glyphicon glyphicon-thumbs-up', # 아이콘 제공 사이트
        prefix='glyphicon'
    )
).add_to(map)
folium.Marker(
    location=[37.30337225,127.0096118],
    popup=folium.Popup('경기도 수원시 장안구 송원로 101', max_width=200),
    tooltip='장안구청',
    icon=folium.Icon(color='green')
).add_to(map)
folium.Marker(
    location=[37.28256171,127.0190196],
    popup=folium.Popup('경기도 수원시 팔달구 창룡대로 23', max_width=200),
    tooltip='팔달구청',
    icon=folium.Icon(icon="cloud",color='red')
).add_to(map)
folium.Circle(
    radius=300, color='crimson',fill=False,
    location=[37.25764489,126.9719386],
    popup=folium.Popup('경기도 수원시 권선구 호매실로 12', max_width=200),
    tooltip='권선구청'
).add_to(map)
folium.CircleMarker(
    radius=10,color="#3186cc",
    fill=True,fill_color="#3186cc",
    location=[37.2596127,127.0465469],
    popup=folium.Popup('경기도 수원시 영통구 효원로 407', max_width=200),
    tooltip='영통구청'
).add_to(map)
title_html='<h3 align="center" style="font-size:20px">수원소재 주요 관공서</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
