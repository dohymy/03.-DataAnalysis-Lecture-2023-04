```python
import requests, json
from urllib.parse import quote
import folium
import pandas as pd
import numpy as np
```

```python
filename = 'data/서울시인구.txt'
pop = pd.read_csv(filename, sep='\t', thousands=',')
pop.head()
```

```python
pop.drop([0, 1, 2], inplace=True)
pop.head()
```

```python
pop=pop[['자치구','인구']]
pop.head()
```

```python
filename = 'data/seoul_geo_simple.json'
geo_data = json.load(open(filename, encoding='utf-8'))
```

```python
area = pd.read_csv('data/서울공원요약.csv')
area.head()
```

```python
area['자치구'] = area.주소.apply(lambda x: ''.join(x.split()[1]))

area.head()
```

```python
sort_area = area[['면적','자치구']].groupby('자치구').sum()
sort_area
```

```python
pop.set_index('자치구', inplace=True)
pop.head()
```

```python
sort_pop=pop.sort_index()
sort_pop
```

```python
pop_area = pd.merge(left = sort_pop , right = sort_area, how = "right", on = "자치구")
pop_area
```

```python
pop_area.drop(['공릉2동', '과천시'],inplace=True)
pop_area
```

```python
pop_area.dtypes
```

```python
pop_area.인구 = pop_area.인구.str.replace(',', '')
pop_area
```

```python
pop_area.인구 = pop_area.인구.astype('int')
```

```python
pop_area.dtypes
```

```python
pop_area['인당 공원면적'] = pop_area['면적']/pop_area['인구']
pop_area
```

```python
map = folium.Map([37.4,127.1], zoom_start=9, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_data,
    data=pop_area['인당 공원면적'],
    columns=[area.index, pop_area['인당 공원면적']],
    fill_color='YlOrRd',
    key_on='properties.name'
).add_to(map)
map
```

```python
area.drop('금천구')
```

```python
def get_text_location(geo_str):
    gu_dict = {}
    for gu in geo_str['features']:
        for coord in gu['geometry']['coordinates']:
            geo = np.array(coord)
            gu_dict[gu['id']] = [np.mean(geo[:,1]), np.mean(geo[:,0])]
    return gu_dict
```

```python
gu_dict = get_text_location(geo_data)
```

```python
map = folium.Map([37.4,127.1], zoom_start=9, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_data,
    data=pop_area['인당 공원면적'],
    columns=[area.index, pop_area['인당 공원면적']],
    fill_color='YlOrRd',
    key_on='properties.name'
).add_to(map)
map
for gu_name in area.index:
    folium.map.Marker(
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
            html=f'<div style="font-size: 10pt">{gu_name}</div>'
        )
    ).add_to(map)
map
```
