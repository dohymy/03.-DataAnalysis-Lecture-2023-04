### 단계 구분도(Cholopleth)
- 서울시 구별 CCTV 설치대수

```python
import json, folium
import pandas as pd
import numpy as np
```

- 데이터 준비

```python
df = pd.read_csv('../02.Pandas/data/cctv2.csv', index_col='구별')
df.head()
```

```python
file = 'data/seoul_geo_simple.json'
geo_data = json.load(open(file, encoding='utf-8'))
```

- 단계구분도

```python
map = folium.Map([37.55,126.98], zoom_start=11, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_data,
    data=df['CCTV 댓수'],       # 단계구분도로 보여줄 데이터
    columns=[df.index, df['CCTV 댓수']],     # 데이터프레임에서 추출할 항목
    fill_color='PuRd',
    key_on='feature.id'     # 지도에 조인할 항목
).add_to(map)
map
```

- 자치구 이름 쓰기

```python
import numpy as np
def get_text_location(geo_str):
    gu_dict = {}
    for gu in geo_str['features']:
        for coord in gu['geometry']['coordinates']:
            geo = np.array(coord)
            gu_dict[gu['id']] = [np.mean(geo[:,1]), np.mean(geo[:,0])]
    return gu_dict
```

```python
gu_dict = get_text_location(geo_data)
gu_dict
```

```python
for gu_name in df.index:
    folium.map.Marker(
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
            html=f'<div style="font-size: 10pt">{gu_name}</div>'
        )
    ).add_to(map)
map
```

- 합치기

```python
map = folium.Map([37.55,126.98], zoom_start=11, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_data,
    data=df['CCTV 댓수'],
    columns=[df.index, df['CCTV 댓수']],
    fill_color='YlOrRd',
    key_on='feature.id'
).add_to(map)
map
for gu_name in df.index:
    folium.map.Marker(
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
            html=f'<div style="font-size: 10pt">{gu_name}</div>'
        )
    ).add_to(map)
title_html='<h3 align="center" style="font-size:20px">서울시 자치구별 CCTV설치댓수</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```

- 최근 증가율로 단계구분도 만들기

```python
map = folium.Map([37.55,126.98], zoom_start=11, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_data,
    data=df['최근 증가율'],
    columns=[df.index, df['최근 증가율']],
    fill_color='YlGn',
    key_on='feature.id'
).add_to(map)
map
for gu_name in df.index:
    folium.map.Marker(
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
            html=f'<div style="font-size: 10pt">{gu_name}</div>'
        )
    ).add_to(map)
title_html='<h3 align="center" style="font-size:20px">서울시 자치구별 CCTV 최근 증가율</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
